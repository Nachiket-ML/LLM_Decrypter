```python
#kobby's method

import Levenshtein 
def calculate_nl(true_text, pred_text):
    dist = Levenshtein.distance(true_text, pred_text)
    max_len = max(len(true_text), len(pred_text))
    if max_len == 0:
        return 1.0 if dist == 0 else 0.0
    nl_score = 1.0 - (dist / max_len)
    return nl_score
```

```python
cipher_method_no_space = []



with open('cipher_method_no_space.json', 'r') as file:
    cipher_method_no_space= json.load(file)
```

```python
distance = 0

for i in cipher_method_no_space:
    truth = cipher_method_no_space[i]['ground_truth'].lower()
    predicted = cipher_method_no_space[i]['predicted'].lower()
    distance += calculate_nl(truth, predicted)
```

```python
distance/len(cipher_method_no_space)
```

```python
cipher_method_with_space = []
with open('cipher_method_with_space.json', 'r', encoding="utf-8") as file:
    cipher_method_with_space= json.load(file)
```
